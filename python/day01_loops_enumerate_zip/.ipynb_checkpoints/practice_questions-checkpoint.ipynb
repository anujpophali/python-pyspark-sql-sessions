## Q1 — 🟢 Easy
### Inventory Status Report

---

### Problem Statement

You are given two lists of equal length:
- `products` — product names
- `stock` — current stock quantity for each product

Using **`enumerate`**, loop over both and build an inventory report as a **dictionary**.  
For each product, label the stock status as:
- `"OUT OF STOCK"` if stock is `0`
- `"LOW"` if stock is between `1` and `10` (inclusive)
- `"OK"` if stock is greater than `10`

---

### Input

```python
products = ["Laptop", "Mouse", "Keyboard", "Monitor", "Webcam", "Headset"]
stock    = [0, 5, 42, 3, 0, 17]
```

---

### Expected Output

```python
result = {
    'inventory_report': [
        {'index': 1, 'product': 'Laptop',   'qty': 0,  'status': 'OUT OF STOCK'},
        {'index': 2, 'product': 'Mouse',    'qty': 5,  'status': 'LOW'},
        {'index': 3, 'product': 'Keyboard', 'qty': 42, 'status': 'OK'},
        {'index': 4, 'product': 'Monitor',  'qty': 3,  'status': 'LOW'},
        {'index': 5, 'product': 'Webcam',   'qty': 0,  'status': 'OUT OF STOCK'},
        {'index': 6, 'product': 'Headset',  'qty': 17, 'status': 'OK'},
    ],
    'summary': {
        'total_products': 6,
        'out_of_stock': 2,
        'low': 2,
        'ok': 2
    }
}
```

---

### Constraints
- Must use `enumerate` (not `range(len(...))`) to generate the row number
- Row numbers are 1-based: `index` starts at 1
- Return a dict with exactly two keys: `'inventory_report'` (list of dicts) and `'summary'` (dict of counts)

````python
# Q1 — Your solution here
import json

products = ["Laptop", "Mouse", "Keyboard", "Monitor", "Webcam", "Headset"]
stock = [0, 5, 42, 3, 0, 17]

results = {'inventory_report': list(), 'summary': dict()}

os, l, ok = 0, 0, 0

for i, (product, qty) in enumerate(zip(products, stock), start=1):

    if qty == 0:
        status = "OUT OF STOCK"
        os += 1
    elif 1 <= qty <= 10:
        status = "LOW"
        l += 1
    else:
        status = "OK"
        ok += 1

    results['inventory_report'].append({
        'index': i,
        'product': product,
        'qty': qty,
        'status': status
    })

results['summary'] = {
    'total_products': len(products),
    'out_of_stock': os,
    'low': l,
    'ok': ok
}

print(json.dumps(results, indent=4))
````

## Q2 — 🟡 Medium
### Payroll Calculator with Bad Data

---

### Problem Statement

You are given three lists:
- `employees` — employee names
- `salaries` — monthly salary (some entries are `None` — missing/corrupt data)
- `months_worked` — number of months worked this year

Using **`zip`** and **`continue`**:
1. Pair all three lists together using `zip`
2. **Skip** any employee whose salary is `None` (use `continue`) — add their name to a `skipped` list
3. For valid employees, compute `annual_pay = salary * months_worked`
4. Return a dict with keys `'payroll'` (list of dicts) and `'summary'` (totals + skipped names)

---

### Input

```python
employees     = ["Alice", "Bob", "Carol", "Dave", "Eve", "Frank"]
salaries      = [8000,    None,  7500,    None,   9200,  6800]
months_worked = [12,      10,    11,      12,     8,     12]
```

---

### Expected Output

```python
result = {
    'payroll': [
        {'name': 'Alice', 'months': 12, 'annual_pay': 96000},
        {'name': 'Carol', 'months': 11, 'annual_pay': 82500},
        {'name': 'Eve',   'months': 8,  'annual_pay': 73600},
        {'name': 'Frank', 'months': 12, 'annual_pay': 81600},
    ],
    'summary': {
        'total_payroll': 333700,
        'processed': 4,
        'skipped': ['Bob', 'Dave']
    }
}
```

---

### Constraints
- Must use `zip` to pair the three lists
- Must use `continue` to skip `None` salaries — do NOT filter the list first
- Return a dict with exactly two keys: `'payroll'` (list of dicts) and `'summary'` (dict)

````python
# Q2 — Your solution here
import json

employees     = ["Alice", "Bob", "Carol", "Dave", "Eve", "Frank"]
salaries      = [8000,    None,  7500,    None,   9200,  6800]
months_worked = [12,      10,    11,      12,     8,     12]


skipped = list()

result = {'payroll': list(), 'summary':dict()}

tot, processed = 0, 0

for i, j in enumerate(zip(employees, salaries, months_worked), start=1):
    annual_pay=0
    if j[1] == None:
        skipped.append(j[0])
        continue
    annual_pay += j[1]*j[2]
    processed += 1
    tot += annual_pay
    result['payroll'].append({'name':j[0], 'months':j[2], 'annual_pay':annual_pay})

result['summary'] = {'total_payroll':tot, 'processed':processed, 'skipped':skipped}

print(json.dumps(result, indent=4))
````

## Q3 — 🔴 Hard
### Pipeline Batch Processor — First Failing Batch

---

### Problem Statement

A data pipeline processes batches of records. You have:
- `batch_ids` — list of batch identifiers
- `records` — list of lists; each inner list contains integer record values for that batch  
  (a record value of `-1` means the record is **corrupted**)

Using **`enumerate`**, **`zip`**, **`continue`**, and **`break`**:

1. Use `zip` to pair `batch_ids` with `records`
2. Use `enumerate` to track the batch number (1-based)
3. For each batch:
   - Loop through records, **skip** corrupted ones (`-1`) using `continue`, and sum the valid ones
   - A batch **fails** if more than **40%** of its records are corrupted
   - If a batch fails, stop processing (`break`) — do not process remaining batches
4. Return a dict with keys `'batches'` (list of batch result dicts) and `'pipeline'` (summary)

---

### Input

```python
batch_ids = ["BATCH-001", "BATCH-002", "BATCH-003", "BATCH-004"]
records   = [
    [120, 340, -1, 210, 180],          # BATCH-001: 1 corrupted out of 5 = 20%
    [500, -1, 430, 290, -1, 310],      # BATCH-002: 2 corrupted out of 6 = 33%
    [100, -1, -1, 200, -1, 150, 180],  # BATCH-003: 3 corrupted out of 7 = 43%  ← FAILS
    [400, 380, 420, 300],              # BATCH-004: should not be processed
]
```

---

### Expected Output

```python
result = {
    'batches': [
        {'batch_num': 1, 'batch_id': 'BATCH-001', 'total': 5, 'corrupted': 1, 'corrupt_pct': 20.0, 'valid_sum': 850,  'status': 'PASS'},
        {'batch_num': 2, 'batch_id': 'BATCH-002', 'total': 6, 'corrupted': 2, 'corrupt_pct': 33.3, 'valid_sum': 1530, 'status': 'PASS'},
        {'batch_num': 3, 'batch_id': 'BATCH-003', 'total': 7, 'corrupted': 3, 'corrupt_pct': 42.9, 'valid_sum': 630,  'status': 'FAIL'},
    ],
    'pipeline': {
        'status': 'FAILED',
        'failed_at_batch': 3,
        'failed_batch_id': 'BATCH-003',
        'batches_processed': 3,
        'batches_total': 4,
        'batches_skipped': 1
    }
}
```

---

### Constraints
- Must use `zip` to pair batch_ids and records
- Must use `enumerate` for the 1-based batch number
- Must use `continue` inside the inner loop to skip `-1` records
- Must use `break` (with `for/else`) to stop on failure
- `corrupt_pct` rounded to 1 decimal place
- Return a dict with exactly two keys: `'batches'` (list of dicts) and `'pipeline'` (summary dict)

````python
# Q3 — Your solution here
import json

batch_ids = ["BATCH-001", "BATCH-002", "BATCH-003", "BATCH-004"]
records   = [
    [120, 340, -1, 210, 180],
    [500, -1, 430, 290, -1, 310],
    [100, -1, -1, 200, -1, 150, 180],
    [400, 380, 420, 300],
]

res = {'batches':list(), 'pipeline':dict()}

s, fb, fbid, bp, bt, bs = '', '', 0, 0, 0, 0

for i, (j, k) in enumerate(zip(batch_ids, records), start=1):
    c, s, sts = 0, 0, ''
    for x in k:
        if x == -1:
            c += 1
            continue
        s += x
    cp = round((c/len(k))*100.0, 1)
    if cp>40.0:
        sts = 'FAIL'
        s = 'FAILED'
    else:
        sts = 'PASS'
        s = 'PASSED'
    res['batches'].append({
        'batch_num':i,
        'batch_id':j,
        'total':len(k),
        'corrupted':c,
        'corrupt_pct':cp,
        'valid_sum':s,
        'status':sts
    })
    if sts == 'FAIL':
        fb = i
        fbid = j
        bp = len(res['batches'])
        bt = len(batch_ids)
        bs = bt-bp
        break

res['pipeline'] = {
    'status': s,
    'failed_at_batch': fb,
    'failed_batch_id': fbid,
    'batches_processed': bp,
    'batches_total': bt,
    'batches_skipped': bs
}
    
print(json.dumps(res, indent=4))
````
